```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("bushelsPerAcre.csv")
df1
```

```python
df1 = df1[df1['Period'] == 'YEAR']
df1
```

```python
df1.rename(columns={'Value': 'Bushels Per Acre'}, inplace=True)
```

```python
yield_df = df1[['Year','Bushels Per Acre']]
yield_df
```

```python
weather_df = pd.read_csv("weather.csv")
weather_df = weather_df[['DATE','PRCP','TAVG','TMAX','TMIN']]
weather_df.rename(columns={'DATE' : 'Year'}, inplace = True)
weather_df
```

```python
corn_yield = pd.merge(yield_df, weather_df, on='Year', how='inner')
corn_yield
```

```python
corn_yield.to_csv("corn_yield.csv")
```
