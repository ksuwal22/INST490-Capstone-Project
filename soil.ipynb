```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/soil.csv")
df
```

```python
df = df[df['Year'].between(2013, 2022)]
df
```

```python
df = df[['Year', 'Data Item', 'Value']]
df
```

```python
averages = df.groupby(['Year', 'Data Item'])['Value'].mean().reset_index()
averages
```

```python
averages[['Soil Type', 'Moisture Category']] = averages['Data Item'].str.split(' - MOISTURE, MEASURED IN PCT ', expand=True)
averages.drop(columns='Data Item', inplace=True)
averages['Soil Type'] = averages['Soil Type'].str.replace('SOIL, ', '')
# Display the modified DataFrame
averages
```

```python
s = averages[averages['Soil Type'] == 'SUBSOIL']
t = averages[averages['Soil Type'] == 'TOPSOIL']
```

```python
pivot_df = s.pivot_table(values='Value', index='Year', columns='Moisture Category', aggfunc='sum')

# Plotting stacked bar graph
pivot_df.plot(kind='bar', stacked=True, color=['skyblue', 'salmon', 'purple', 'green'], figsize=(10, 6))

plt.title('Moisture Category Comparison for Subsoil (2013-2022)')
plt.xlabel('Moisture Category')
plt.ylabel('(%)')
plt.legend(title='Soil Type')

plt.show()
```

```python
pivot_df = t.pivot_table(values='Value', index='Year', columns='Moisture Category', aggfunc='sum')

# Plotting stacked bar graph
pivot_df.plot(kind='bar', stacked=True, color=['skyblue', 'salmon', 'purple', 'green'], figsize=(10, 6))

plt.title('Moisture Category Comparison for Topsoil (2013-2022)')
plt.xlabel('Moisture Category')
plt.ylabel('(%)')
plt.legend(title='Soil Type')

plt.show()
```

```python
df = averages[averages['Moisture Category'] == 'ADEQUATE']
fig, ax = plt.subplots(figsize=(10, 6))

for soil_type in df['Soil Type'].unique():
    for moisture_category in df['Moisture Category'].unique():
        subset = df[(df['Soil Type'] == soil_type) & (df['Moisture Category'] == moisture_category)]
        ax.plot(subset['Year'], subset['Value'], marker='o', label=f'{soil_type} - {moisture_category}')

plt.title('Value Trend over Years by Adequate Soil Type')
plt.xlabel('Year')
plt.ylabel('Adequate Soil (%)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
dff = pd.read_csv('/content/corn_yield (1).csv')
dff = dff[dff['Year'].between(1995, 2022)]
dff
```

```python
original = pd.read_csv("/content/soil.csv")
original = original[original['Year'].between(1995, 2022)]
original = original[['Year', 'Data Item', 'Value']]
original = original.groupby(['Year', 'Data Item'])['Value'].mean().reset_index()
original[['Soil Type', 'Moisture Category']] = original['Data Item'].str.split(' - MOISTURE, MEASURED IN PCT ', expand=True)
original.drop(columns='Data Item', inplace=True)
original['Soil Type'] = original['Soil Type'].str.replace('SOIL, ', '')
original = original[original['Moisture Category'] == 'ADEQUATE']
original = original.groupby(['Year', 'Soil Type'])['Value'].mean().unstack().reset_index()

original
```

```python
merged_df = pd.merge(dff, original, on='Year')
df = merged_df
```

```python
correlation = df['SUBSOIL'].corr(df['Bushels Per Acre'])
print(f"Correlation between Adequate Subsoil and Yield: {correlation:.2f}")

# Scatter plot to visualize the relationship
plt.figure(figsize=(8, 6))
plt.scatter(df['SUBSOIL'], df['Bushels Per Acre'])
plt.title('Adequate Subsoil vs. Corn Yield')
plt.xlabel('Adequate Subsoil (%)')
plt.ylabel('Corn Yield (Bushels Per Acre)')
plt.grid(True)
plt.show()
```

```python
correlation = df['TOPSOIL'].corr(df['Bushels Per Acre'])
print(f"Correlation between Adequate Subsoil and Yield: {correlation:.2f}")

# Scatter plot to visualize the relationship
plt.figure(figsize=(8, 6))
plt.scatter(df['TOPSOIL'], df['Bushels Per Acre'])
plt.title('Adequate Subsoil vs. Corn Yield')
plt.xlabel('Adequate Subsoil (%)')
plt.ylabel('Corn Yield (Bushels Per Acre)')
plt.grid(True)
plt.show()
```

```python
correlation = df['PRCP'].corr(df['SUBSOIL'])
print(f"Correlation between Precipitation and Adequate Subsoil: {correlation:.2f}")

# Scatter plot to visualize the relationship
plt.figure(figsize=(8, 6))
plt.scatter(df['PRCP'], df['SUBSOIL'])
plt.title('Precipitation vs. Adequate Subsoil')
plt.xlabel('Precipitation (in.)')
plt.ylabel('Adequate Subsoil (%)')
plt.grid(True)
plt.show()
```

```python
correlation = df['TAVG'].corr(df['SUBSOIL'])
print(f"Correlation between Average Temperature and Adequate Subsoil: {correlation:.2f}")

# Scatter plot to visualize the relationship
plt.figure(figsize=(8, 6))
plt.scatter(df['TAVG'], df['SUBSOIL'])
plt.title('Average Temperature vs. Adequate Subsoil')
plt.xlabel('Average Temperature (K)')
plt.ylabel('Adequate Subsoil (%)')
plt.grid(True)
plt.show()
```

```python
import statsmodels.api as sm
df['Constant'] = 1

# Performing linear regression
model = sm.OLS(df['Bushels Per Acre'], df[['SUBSOIL', 'Constant']])
results = model.fit()

# Visualizing the regression line and data points
plt.figure(figsize=(8, 6))
plt.scatter(df['SUBSOIL'], df['Bushels Per Acre'], label='Actual Data')
plt.plot(df['SUBSOIL'], results.predict(), color='red', label='Regression Line')
plt.title('Adequate Subsoil vs. Corn Yield with Regression Line')
plt.xlabel('Adequate Subsoil (%)')
plt.ylabel('Corn Yield (Bushels Per Acre)')
plt.legend()
plt.grid(True)
plt.show()
```
